```python
import pandas as pd
```

```python
import pickle
with open('summary.pickle', 'rb') as f:
    summary = pickle.load(f) 
```

```python
#count값 작은것 부터 나열
sort_summary =summary.sort_values(by=['count'], axis=0)
sort_summary.reset_index(drop=True)
```

```python
#false true 슬라이싱
falseData = sort_summary[10:60]
trueData = sort_summary[1131:]

#labeling
falseData['Label'] = False
trueData['Label']=True

#concat
trainData = pd.concat([falseData,trueData])
```

```python
trainData
```

```python
pickle.dump(trainData, open("trainData.pickle", "wb" ))
```
